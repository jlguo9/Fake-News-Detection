```python
import pandas as pd
```

```python
liar_data = pd.read_csv('liar_data.csv')
pheme_data = pd.read_csv('pheme_data.csv')
buzzfeed_data = pd.read_csv('buzzfeed_data.csv')
```

## Extract Text and Label

### liar_data

```python
liar_data = liar_data[['label','statement']]
# transfer "half_true","mostly_true", etc to "true"
liar_data['label'] = liar_data['label'].replace(to_replace='.*true.*', value='true', regex=True)
liar_data['label'] = liar_data['label'].replace(to_replace='.*liar.*', value='false', regex=True)
liar_data['label'] = pd.get_dummies(liar_data['label'])['false']
liar_data = liar_data.rename(columns={'statement': 'text'})
```

### pheme_data

```python
pheme_data = pheme_data[['label','text']]
```

### buzzfeed_data

```python
# TODO: consider whether merge title into text

buzzfeed_data['label'] = pd.get_dummies(buzzfeed_data['type'])['Fake']
# only use title
buzzfeed_data = buzzfeed_data[['title','label']].rename(columns={'title': 'text'})
```

## Concat Data

```python
data = pd.concat([liar_data, pheme_data, buzzfeed_data], axis=0)
data.head()
```

```python
# save
data.to_csv('data.csv', index=False)
```
